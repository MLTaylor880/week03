##### Write a brief description of your data, and any interesting observations you've made thus far. 

The data is from Billboard's list of songs for 2000 by artist, genre, time, dates, and standing in the charts. 

List includes only songs that peaked in 2000. For each song the weeks (columns) represent the first week it entered the charts. For example, a song that entered the charts July 17, 1999, the first week column contains the ranking it entered July 17, 1999. It peaked February 12, 2000. Week 31 has it's peak ranking. This explains why when you sort the week column from low to high the numbers do not range from 1000.  The weeks are relative to each song.

There are 10 genres in the chart. When producing the playlist for the DJ, top 2 longest in the charts for each genre, jazz, raggae, and gospel genres did not have 2 songs in the charts.

```python
#checking the track (song) column. There are 317 strings and none are blank
print billboard.dtypes['track']
print billboard['track'].count()
print billboard['track'].isnull().sum()
```

```python
#The time column was changed to length since it is the length of the column
#Length was converted to seconds so they can be compared, added etc.
#When the DJ only has so much time to play the music, talk and play ads.
#There 317 items  The shortes song is 2.6 minutes and the longest is 7.8 minutes

def time_splitter_quick(length):
    new_length = length.split(':')
    return 60*int(length[0]) + int(new_length[1])

billboard['playtime_in_seconds'] = billboard['length'].apply(time_splitter_quick)
billboard['playtime_in_seconds'].describe()
```

```python
#Genre is the type of music
#There are 317 items and 10 categories
print billboard.dtypes['genre']
print billboard['genre'].isnull().sum()
print billboard['genre'].count()
lstGenre = [billboard['genre'].drop_duplicates()]
print lstGenre
```

```python
#Date peaked is the week the song peaked in the charts
#There are 317 dates and all are within 2000
print billboard.dtypes['date.peaked']
print billboard['date.peaked'].isnull().sum()
billboard['date.peaked']=pd.to_datetime(billboard['date.peaked'],format='%Y-%m-%d')
print billboard['date.peaked'].describe()
```

```python
#Looking at all the entries in the weeks. A list was created to hold all 
#entries so the could be processed in a loop. Important to note there is
#missing data.  It was decided that it was not worth the time to fill
#in all the missing cells with 0.  It this had been a date that would 
#affect more than one's listening pleasure then the cells would have
#been filled in
#Notice the first week the is 317 after that the count drops off. Songs
#drop off the chart or there is sloppy input.
lstWeeks = []
i=0
for labels in billboard.columns:  
    if i >6:
        lstWeeks.append(labels)
        i = i+1
    else:
        i=i+1

for item in lstWeeks:
    print 'For %s' %item
    print billboard.dtypes[item] 
    print billboard[item].isnull().sum()
    print billboard[item].describe()
    print ' '
```

##### Using Pandas' built in `melt` function, pivot the weekly ranking data to be long rather than wide.
```python
#Decision made to drop any nulls
billboardMelt =billboardMelt.dropna()
billboardMelt.head(5)
```

```python
#This data will be used in Tableau
rankingData = pd.pivot_table(billboardMelt,
                           index = ['artist', 'track','length','genre',],
                           values = ['Rank'],aggfunc=[np.min, np.mean, len])

rankingData
```
