```python
%matplotlib inline
import random
import numpy as np
from matplotlib import pyplot as plt
from sklearn import datasets, linear_model
plt.rcParams['figure.figsize'] = 8, 8

# Generate some data
def generate_data():
    xs = np.arange(-5, 5, 1)
    #puts in a numpy array vs a list
    
    data = [(x - random.random(), (x + random.random())**2) for x in xs]
    data.sort()
    xs = [x for (x, y) in data]
    ys = [y for (x, y) in data]
    return xs, ys
#created an array and then converted it to a list
xs, ys = generate_data()
plt.scatter(xs, ys)
plt.title("Randomly Generated Data")
plt.show()
```

```python
type(xs)
```

```python
lm = linear_model.LinearRegression()
```

```python
# This function from numpy builds a matrix of powers for us
X = np.vander(xs, 4)
#give me a row each of the powers of 4 column 1 is power to the 3 or x^3
y = ys
print X,y
```

```python
model = lm.fit(X, y)
predictions = lm.predict(X)
print predictions
for item in  zip(model.coef__,['constant','x','x^2','x^3']);
print item
```

```python

plt.scatter(xs, ys)
plt.title("Randomly Generated Data")
plt.plot(xs, predictions)
plt.show()
print "r^2:", model.score(X, y)
```

```python
xs2, ys2 = generate_data()
X = np.vander(xs2, 4)
#new set of x and y
predictions = lm.predict(X)

plt.scatter(xs2, ys2)
plt.title("Randomly Generated Dataset #2")
plt.plot(xs2, predictions)
plt.show()
print "r^2:", model.score(X, ys2)
```

```python
# Note: alpha plays the role of lambda in sklearn (lambda is the notation on e.g. Wikipedia)
rlm = linear_model.Ridge(alpha=4, normalize=True)
#alpha 4 is really big normalizing true wouldn't compare data if normalize data

# Fit the polynomial again with ridge regularization
X = np.vander(xs, 4)
y = ys
ridge_model = rlm.fit(X, y)
predictions = ridge_model.predict(X)

plt.scatter(xs, ys)
plt.title("Randomly Generated Data")
plt.plot(xs, predictions)
plt.show()
print "r^2:", ridge_model.score(X, ys)
```
